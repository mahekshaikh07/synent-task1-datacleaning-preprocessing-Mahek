## Step 1: Load the Dataset

```python
df = pd.read_csv('titanic.csv')
```

## Step 2: Inspect the Data

```python
df.info()
```

```python
df.isnull().sum()
```

## Step 3: Handle Missing Values
'Age' - fill with median 
'Embarked' - fill with mode
'Cabin' - drop

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(columns=['Cabin'], inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

## Step 4: Remove Duplicates

```python
print("Duplicates before: ", df.duplicated().sum())
df.drop_duplicates(inplace = True)
```

```python
print("Duplicates after: ", df.duplicated().sum())
```

## Step 5: Convert Data Types
`Survived` and `Pclass` should be categorical, not integers

```python
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
print(df.dtypes)
```

## Step 6: Rename Columns
Making column names lowercase and more readable

```python
df.rename(columns={
    'PassengerId': 'passenger_id',
    'Survived': 'survived',
    'Pclass': 'pclass',
    'Name': 'name',
    'Sex': 'sex',
    'Age': 'age',
    'SibSp': 'siblings_spouses',
    'Parch': 'parents_children',
    'Ticket': 'ticket',
    'Fare': 'fare',
    'Embarked': 'embarked'
}, inplace=True)

df.head()
```

## Step 7: Save Cleaned dataset

```python
df.to_csv('titanic_cleaned.csv', index=False)
print("Cleaned dataset saved as titanic_cleaned.csv")
print("Final shape:", df.shape)
```

```python
## Summary
| Step | Action | Result |
|------|--------|--------|
| Missing Values | Filled Age with median, Embarked with mode, dropped Cabin | No nulls remaining |
| Duplicates | Checked and removed | 0 duplicates |
| Data Types | Survived, Pclass → category | Correct types |
| Columns | Renamed to lowercase | Clean naming convention |

Dataset is now clean and ready for analysis.
```
